```python
import cv2
from matplotlib import pyplot as plt

# Load the cascade classifier XML file for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
# Read the image
image = cv2.imread('Artificial.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Perform face detection
faces = face_cascade.detectMultiScale2(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Unpack the result
rectangles, weights = faces

# Draw rectangles around the detected faces
for (x, y, w, h) in rectangles:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

# Display the image with detected faces
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2GRAY))
plt.title(" face detection")
```

```python
edges = cv2.Canny(gray, 50, 500)

plt.imshow(cv2.cvtColor(edges,cv2.COLOR_BGR2RGB))
plt.title("edge detection image")
```
